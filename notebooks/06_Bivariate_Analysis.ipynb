### Objective

Analyze the relationship between customer characteristics and churn to identify the factors that are most strongly associated with customer attrition.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (9,5)
```

```python
df = pd.read_csv(
"/content/drive/MyDrive/ML Projects/End-to-End-Customer-Churn-ML-Pipeline/data/raw/customer_churn.csv"
)
```

```python
sns.countplot(x="Contract", data=df, hue="Churn")
plt.show()
```

```python
pd.crosstab(df["Contract"], df["Churn"])
```

```python
pd.crosstab(df["Contract"], df["Churn"], normalize="index")*100
```

```python
sns.countplot(x="Contract", data=df, hue="Churn")

plt.title("Contract Type vs Churn")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")

plt.savefig(
"/content/drive/MyDrive/ML Projects/End-to-End-Customer-Churn-ML-Pipeline/images/contract_vs_churn.png",
dpi=300,
bbox_inches="tight"
)

plt.show()
```

- Observation  
We can see that month-to-month contracts have a larger churn percentage than 1 and 2 year contracts.

- Business Interpretation  
This observation indicates that a large proportion of customers are on month-to-month contracts, indicating greater flexibility but potentially lower customer commitment. This group also exhibits the highest churn.

- Business Recommendation  
Appropriate strategies need to be developed to promote customers to join the long-term contracts. Introduction of targeted incentives could encourage long-term contracts therefore increasing customer retention
