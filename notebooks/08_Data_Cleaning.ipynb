# End-to-End Customer Churn ML Pipeline

## Notebook 08 – Data Cleaning

### Objective

Prepare the raw customer churn dataset for machine learning by identifying and correcting data quality issues, handling missing values, converting data types, removing unnecessary columns, and validating the cleaned dataset.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    "/content/drive/MyDrive/ML Projects/End-to-End-Customer-Churn-ML-Pipeline/data/raw/customer_churn.csv"
)
```

```python
df.shape
```

```python
df.dtypes
```

```python
(df['TotalCharges']==" ").sum()
```

```python
df[df['TotalCharges']==" "]
```

Tenure = 0. This implies that these are new customers who havent accumulated any charges yet.

```python
df["TotalCharges"]=df['TotalCharges'].replace(" ",np.nan)
```

```python
df.isnull().sum()
```

```python
(df['TotalCharges']==" ").sum()
```

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'], errors="coerce")
```

```python
df.info()
```

Since only small number of rows have missing values, their removal wont effect the model. Therefore those rows can be dropped.

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

No duplicates are present hence no need to remove duplicates.

```python
df=df.drop("customerID", axis=1)
```

```python
df.to_csv(
    "/content/drive/MyDrive/ML Projects/End-to-End-Customer-Churn-ML-Pipeline/data/processed/customer_churn_cleaned.csv",
    index=False
)
```

```python
missing_percentage=(
    df.isnull().sum()/len(df)
) * 100

missing_percentage
```
