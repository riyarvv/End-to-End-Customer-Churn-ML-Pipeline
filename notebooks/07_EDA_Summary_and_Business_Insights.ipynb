## Executive Summary

1. Business Problem  
This project analyzes customer churn using IBM Telco Customer Churn data. It aims to find customer charactaristics and devise strategies to reduce customer churn. This is highly relevant for most telecommunication companies as retaining existing customers is more viable than acquiring new ones.

2. Dataset Overview  
It contains 7043 customer informations across demographic details, subscribed services, billing information, contract types, and churn status. Both categorical and numerical variables were explored to understand customer behavior before building any machine learning models.

3. Customer Demographics  
Customers with partners and dependents generally exhibited lower churn rates, suggesting that household or family-related factors may contribute to greater customer stability.

4. Contract & Tenure  
Contract type emerged as one of the strongest indicators of churn. Customers on month-to-month contracts were substantially more likely to leave than customers with one-year or two-year contracts. Similarly, customers with shorter tenure tended to churn more frequently.

5. Services
Customers without online security or technical support experienced higher churn rates. Additionally, fiber optic customers appeared to churn more often than DSL customers, suggesting that service quality, pricing, or customer expectations may influence retention.

6. Billing  
Customers using electronic checks and paperless billing displayed relatively higher churn. While these variables alone do not explain customer behavior, they may help identify customer segments that warrant further investigation.

7. Numerical Analysis  
Higher monthly charges appeared to be associated with increased churn, whereas longer-tenure customers generally remained with the company.
