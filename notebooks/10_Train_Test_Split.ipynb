1. Why do we use stratify=y?  
To ensure that our training and testing datasets maintain the exact same proportion of target classes as the original dataset,

2. What is the purpose of X and y?  
X = input features
y = target variable

3. Why is data leakage dangerous?  
It secretly taints the training process by letting the model "cheat" or peek at the answers.

4. Why do we set random_state=42?  
Consistent results and fair comparison.
