```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")

plt.rcParams["figure.figsize"] = (8,5)
```

```python
df = pd.read_csv(
    "/content/drive/MyDrive/ML Projects/End-to-End-Customer-Churn-ML-Pipeline/data/raw/customer_churn.csv"
)
```

```python
df['InternetService'].value_counts()
```

```python
sns.countplot(x='InternetService',
              data=df,
              order=df["InternetService"].value_counts().index
)
plt.title('Internet Service Distribution')
plt.xlabel('Internet Service')
plt.ylabel('Number of Customers')
plt.savefig(
    "/content/drive/MyDrive/ML Projects/End-to-End-Customer-Churn-ML-Pipeline/images/internet_service_distribution.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
plt.close()
```

Fiber optic is the most widely used internet service.
More customers use internet service options than those who dont.
