## Notebook 03 - Exploratory Data Analysis

### Objective
Explore the dataset using visualizations to understand customer characteristics and identify patterns related to customer churn.


```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/ML Projects/End-to-End-Customer-Churn-ML-Pipeline/data/raw/customer_churn.csv")
```

```python
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (8,5)
```

```python
df['Churn'].value_counts()
```

```python
df['Churn'].value_counts(normalize=True)*100
```

```python
sns.countplot(x='Churn', data=df)
plt.title('Customer Churn Distribution')
plt.savefig(
    "/content/drive/MyDrive/ML Projects/End-to-End-Customer-Churn-ML-Pipeline/images/churn_distribution.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

From this plot we can infer that lesser number of customers have churned.
